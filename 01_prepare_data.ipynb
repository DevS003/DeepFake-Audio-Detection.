### **Step 1 — Import libraries**

```python
import os # work with folders & paths
import shutil # copy files
from tqdm import tqdm # progress bar (VERY important, dataset has thousands of files)
```

### **Step 2 — Set project paths**

```python
import os

BASE_PATH = r"D:\Deepfake_Audio_Project"

RAW_LA = os.path.join(BASE_PATH, "raw_data", "LA")
CLEAN = os.path.join(BASE_PATH, "clean_data")

print("RAW_LA path:", RAW_LA)
print("CLEAN path:", CLEAN)
```

### **Step 3 — Create output folders**

```python
splits = ["train", "dev", "eval"]
labels = ["real", "fake"]

for s in splits:
    for l in labels:
        os.makedirs(os.path.join(CLEAN, s, l), exist_ok=True)

print("Folders created")
 # create folders for train/dev/eval and real/fake in the clean_data directory
 # exist_ok=True means that if the folder already exists, it won't throw an error
 # os.path.join(clean, s, l) creates the path for each folder (e.g. clean/train/real)
```

### **Step 4 — The Core Function**

```python
def organize_split(protocol_file, audio_folder, output_folder):

    with open(protocol_file, "r") as f:
        lines = f.readlines()

    for line in tqdm(lines):
        parts = line.strip().split()

        file_id = parts[1]       # audio name
        label = parts[-1]        # bonafide or spoof

        source_file = os.path.join(audio_folder, "flac", file_id + ".flac")

        if label == "bonafide":
            dest_file = os.path.join(output_folder, "real", file_id + ".flac")
        else:
            dest_file = os.path.join(output_folder, "fake", file_id + ".flac")

        if os.path.exists(source_file):
            shutil.copy(source_file, dest_file)

```

##### **What the function does overall**

For every audio file:

1.Read label

2.Find audio

3.Check real/fake

4.Copy to correct folder


### **Step 5 — Run for TRAIN**

```python
organize_split(
    os.path.join(RAW_LA, "ASVspoof2019_LA_cm_protocols", "ASVspoof2019.LA.cm.train.trn.txt"),
    os.path.join(RAW_LA, "ASVspoof2019_LA_train"),
    os.path.join(CLEAN, "train")
)
```

### **Step 6 — Run for DEV**

```python
organize_split(
    os.path.join(RAW_LA, "ASVspoof2019_LA_cm_protocols", "ASVspoof2019.LA.cm.dev.trl.txt"),
    os.path.join(RAW_LA, "ASVspoof2019_LA_dev"),
    os.path.join(CLEAN, "dev")
)
```

### **Step 7 — Run for EVAL**

```python
organize_split(
    os.path.join(RAW_LA, "ASVspoof2019_LA_cm_protocols", "ASVspoof2019.LA.cm.eval.trl.txt"),
    os.path.join(RAW_LA, "ASVspoof2019_LA_eval"),
    os.path.join(CLEAN, "eval")
)
```
